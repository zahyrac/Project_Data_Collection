```python
from bs4 import BeautifulSoup
import urllib.request
import requests
import csv
import pandas as pd
```

```python
ratinglist=[]
reviewlist=[]

i=0
while True:
    url = f'https://www.yelp.com/biz/code-red-restaurant-and-lounge-bronx-2?start={i}'
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    get_ul_tag = soup.find_all('ul', {'class': 'undefined list__373c0__vNxqp'})
    for ul_tag in get_ul_tag:
        all_li = ul_tag.find_all('li')
        for all_text_rating in all_li:
            ratings = all_text_rating.find('div', {'class': 'i-stars__373c0___sZu0'}).get('aria-label')
            ratinglist.append(ratings)
            reviews = all_text_rating.find('span', {'class': 'raw__373c0__tQAx6'}).text
            reviewlist.append(reviews)
            print(f"{ratings}\n{reviews}\n\n")
    if i>=60:
        break
    i = i+10
```

```python
page_output = pd.DataFrame(
    {'Star Rating': ratinglist,
     'Reviews': reviewlist
    })
    
page_output
    
```
